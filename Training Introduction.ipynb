## Build a Basic Translator
- take a string and translate it

```python
def translate (phrase):
    translation = ""
    for letter in phrase:
        if letter in "AEIOUaeiou":
            if letter.isupper ():
                translation = translation + "G"
            else:
                translation = translation + "g"
        else:
            translation = translation + letter
    return translation

print (translate(input("Enter a phrase: ")))
```

## Building a Multiple Choice Quiz
- using loops, classes, if statements

```python
question_prompts = [
    "What colour are apples?\n(a) Red/Green\n(b) Purple\n(c) Orange\n\n",
    "What colour are bananas?\n(a) Teal\n(b) Magenta\n(c) Yellow\n\n",
    "What colour are strawberries?\n(a) Yellow\n(b) Red\n(c) Blue\n\n",
]

class question:
    def __init__(self, prompt, answer):
        self.prompt = prompt
        self.answer = answer
        
questions = [
    question (question_prompts [0], "a"),
    question (question_prompts [1], "c"),
    question (question_prompts [2], "b"),
]

def run_test (questions):
    score = 0
    for question in questions:
        answer = input (question.prompt)
        if answer == question.answer: 
            score += 1
    print ("You got "+ str(score) + "/" + str(len(questions)) + " correct!")
    
run_test(questions)
```

## Inheritance
- define attributes in a class
- inherit attributes in a new class

```python
class chef:
    def make_chicken (self):
        print ("The chef makes a chicken")
        
    def make_salad (self):
        print ("The chef makes a salad")
        
    def make_special_dish (self):
        print ("The chef makes a special dish: Meatballs")
        
my_chef = chef ()
my_chef.make_special_dish()

class chinese_chef(chef):
        
    def make_special_dish (self):
        print("The chef makes a special dish: Singapore Noodles")
        
    def make_fried_rice (self):
        print ("The chef makes fried rice")

my_chinese_chef = chinese_chef ()
my_chinese_chef.make_special_dish()

```
